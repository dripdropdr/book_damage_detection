```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def plt_time(df):
    
    ax = df.plot(x=["Default", "Default+Aug.", "Default3x", "Default3x+Aug.", "Mosaic3x"], y=None, kind="scatter", figsize=(12, 8))

    # x축, y축 이름 및 범례 설정
    plt.ylabel('Training time (h)', size = 13)
    plt.xticks([0,1,2,3,4], ["Default", "Default+Aug.", "Default3x", "Default3x+Aug.", "Mosaic3x"], rotation=0)
    
    for p in ax.patches:
        ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
    
#     plt.title("YOLOX", pad=20)
    plt.legend()
    plt.show()
```

```python
def plt_cls_draw(cls, df):
    
    ax = df.plot(y=["AP@50", "AP@50:95"], kind="bar", figsize=(12, 8))

    # x축, y축 이름 및 범례 설정
    plt.ylabel('AP', size = 13)
    plt.xticks([0,1,2,3,4,5], ["Barcode", "Notch", "Ripped", "Spot", "Tag", "Wear"], rotation=0)
    
    for p in ax.patches:
        ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
    
#     plt.title("YOLOX", pad=20)
    plt.legend()
    plt.show()
```
